````python
l = [ 0, 1, 2, 3, 5, 6, 7 ]
def find_it(l):
  if l is None: return None
  l2 = []
  for i in range(len(l)):
    if i%2==0:
      l2.append(l[i])
  return l2

assert find_it(None)==None
assert find_it([])==[]
assert find_it([1])==[1]
assert find_it([1,2])==[1]
assert find_it([0,1,2,3,4,5,7])==[0,2,4,7]
````

````python
path_names = os.listdir("annotations")
# path_names = ['README', 'trainval.txt', '._trimaps', 'xmls', 'list.txt', 'trimaps', 'test.txt']

file_list = []

for path in path_names:
  # a_path = "/".join(["annotations", path])
  a_path = "annotations/{}".format(path)
  # a_path = "annotations/README"
  is_file = os.path.isfile(a_path)
  file_type = "file" if is_file else "directory"
  # print("{} : {}".format(path, file_type))

  if is_file:
    file_list.append(a_path)

for a_file in file_list:
  f = open(a_file, "r")
  lines = f.readlines()
  f.close()

  print("file name : {}".format(a_file.split("/")[1]))
  for i in range(5):
    print(lines[i], end="")
  print()
````

````python
def get_file_list(target_directory_name):
  path_names = os.listdir(target_directory_name)
  # path_names = ['README', 'trainval.txt', '._trimaps', 'xmls', 'list.txt', 'trimaps', 'test.txt']

  file_list = []
  for path in path_names:
    a_path = "{}/{}".format(target_directory_name, path)
    is_file = os.path.isfile(a_path)
    file_type = "file" if is_file else "directory"
    if is_file:
      file_list.append(a_path)

  return file_list

assert get_file_list("annotations") == ['annotations/README', 'annotations/trainval.txt', 'annotations/list.txt', 'annotations/test.txt']



def read_file(file_name):
  f = open(file_name, "r")
  lines = f.readlines()
  f.close()
  return lines

def show_head(file_name, lines, line_count):
  print("file name : {}".format(file_name.split("/")[1]))
  for i in range(line_count):
    print(lines[i], end="")
  print()


# task1. annotations 밑의 디렉토리가 아닌 파일 리스트 구하기
file_list = get_file_list("annotations")

# task2. 파일리스트에서 루프를 돌면서
for file_name in file_list:
# task2-1. 파일을 열어서 읽기
  lines = read_file(file_name)
# task2-2. 파일이름과 내용 5줄 출력
  show_head(file_name, lines, 5)
````

````python
file_list = get_file_list("annotations")

for file_name in file_list:
  lines = read_file(file_name)
  show_head(file_name, lines, 5)
````
