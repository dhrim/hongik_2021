## plt.subplot()

plt.plot()의 호출은 내부적으로 plot.hidden_current_axes.plot()을 통해 이루어 진다.

plt.subplot()을 호출하면 sub axes이 하나 생성되고 plot.hidden_current_axes가 새로 생성된 sub axes를 가리키게된다.

그래서 이후에 plt.plot()과 같이 호출하면 새로 생긴 sub axes을 호출해서 그리는 것과 동일하다.

````
x = np.linspace(0, 10, 100)

fig = plt.figure()

ax1 = fig.add_subplot(211)
ax1.plot(x, np.sin(x), '-o')
ax1.plot(x, np.cos(x), '^')

ax2 = fig.add_subplot(212)
ax2.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))

plt.show()


print()


plt.subplot(211)
plt.plot(x, np.sin(x), '-o')
plt.plot(x, np.cos(x), '^')

plt.subplot(212)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
````

fig.add_subplot()은 반환된 axes를 사용하여 그린다.

plt.subplot()은 plt로 그릴 대상을 지정하고 plt를 사용하여 그린다.

````
fig = plt.figure(facecolor="gray")

fig.add_subplot(121, title="ax121")
fig.add_subplot(122, title="ax122")

plt.tight_layout()
plt.show()

print()

fig = plt.figure(facecolor="gray")

ax121 = plt.subplot(121, title="ax121")
ax122 = plt.subplot(122, title="ax122")

plt.tight_layout()
plt.show()
````

### plt.subplot()

````
fig = plt.figure()

plt.subplot(211)
plt.plot(x, np.sin(x))

plt.subplot(212)
plt.plot(x, np.cos(x))

plt.show()
````

### plt.subplots()

````
fig, axes = plt.subplots(2)

axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))

plt.show()
````
