# Quiz

y = x**2 그리기

````
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5,5,0.1)
y = x**2

plt.plot(x, y)
plt.show()
````

# Quiz

y와 실제 출력 y_를 동시에 그리기





````
import numpy as np
import matplotlib.pyplot as plt

def dnn_predict(x):
  return x**2 + np.random.rand(x.shape[0])
  
x = np.arange(-5,5,0.1)
y = x**2
y_ = dnn_predict(x)

plt.plot(y)
plt.plot(y_)
plt.show()
````
