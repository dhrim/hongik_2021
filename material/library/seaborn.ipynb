# regplot() - 선형회귀

copy from https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot

````python
tips = sns.load_dataset("tips")
````

## 기본

````python
ax = sns.regplot(x="total_bill", y="tip", data=tips)
````

## Numpy Array로

````python
import numpy as np; np.random.seed(8)
mean, cov = [4, 6], [(1.5, .7), (.7, 1)]
x, y = np.random.multivariate_normal(mean, cov, 80).T
ax = sns.regplot(x=x, y=y, color="g")
````

## Pandas Series로

````python
import pandas as pd
x, y = pd.Series(x, name="x_var"), pd.Series(y, name="y_var")
ax = sns.regplot(x=x, y=y, marker="+")
````

## 신뢰구간 적용
confidence interval 

![대체 텍스트](https://github.com/dhrim/hongik_2021/raw/master/material/library/images/confidence_interval.png)

image from https://www.youtube.com/watch?v=w3tM-PMThXk

````python
ax = sns.regplot(x=x, y=y, ci=68, truncate=False)
````

## 비연속 데이터

````python
ax = sns.regplot(x="size", y="total_bill", data=tips)
````

x값들에 변동을 줘서

````python
ax = sns.regplot(x="size", y="total_bill", data=tips, x_jitter=.1)
````

평균으로

````python
ax = sns.regplot(x="size", y="total_bill", data=tips, x_estimator=np.mean)
````

````python
ax = sns.regplot(x=x, y=y, x_bins=4)
````

## 다차원 회귀

````python
ans = sns.load_dataset("anscombe")
ax = sns.regplot(x="x", y="y", data=ans.loc[ans.dataset == "II"],
                 scatter_kws={"s": 80},
                 order=2, ci=None)
````

## 로그 스케일

````python
ax = sns.regplot(x="size", y="total_bill", data=tips,
                 x_estimator=np.mean, logx=False)
````

## 연속 데이터의 신뢰구간

````python
fmri = sns.load_dataset("fmri")
````

````python
print(fmri)
````

````python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
````

ci에 default 값 0.95가 적용된 것.

````python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, ci=None)
````

````python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, ci="sd");
````

ci에 표준편차 "sd" 값 설정함.
