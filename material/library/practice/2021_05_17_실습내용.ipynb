````python
from IPython.display import clear_output

MIN = 1
MAX = 19

def draw_of_position(p):
  left_length = p-1
  right_length = 19-p
  figure = "|" + ("-"*left_length) + "x" + ("-"*right_length) +"|"
  print(figure)

def move_position(i, p):
  if i=="q":
    p = max(MIN,p-1)
  elif i=="e":
    p = min(MAX,p+1)
  return p

p = 10
while True:
  i = input()
  if i=="exit": break
  p = move_position(i,p)
  clear_output()
  draw_of_position(p)
````
