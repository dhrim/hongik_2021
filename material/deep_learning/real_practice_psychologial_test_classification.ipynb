# Voted 컬럼 보기

```
df.groupby('voted').count()
```

```
df = df.drop(df[df['voted']==0].index)
```

```
df.groupby('voted').count()
```

# Normalization

```
normalized_df=(df-df.min())/(df.max()-df.min())
```

```
df.isnull().any().any()
```

# X, Y 구하기

```
voted_index = df.columns.get_loc("voted")
print(voted_index)
```

```
data = normalized_df.to_numpy()
print(data.shape)
print(data[0])
```

```
y = data[:,voted_index]
print(y.shape)
```

```
print(y[:10])
```

```
import matplotlib.pyplot as plt
plt.hist(y)
plt.show()
```

```
import numpy as np

print(data.shape)
x = np.delete(data, np.s_[voted_index], axis=1) 
print(x.shape)
```

# train, test 나누기

```
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, train_size=0.7)
```

```
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

# 모델 정의, 학습

```
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense
```

```
from tensorflow.keras.layers import *

model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(train_x.shape[1],)),
    BatchNormalization(),
    keras.layers.Dense(32, activation='relu'),
    BatchNormalization(),
    keras.layers.Dense(2, activation='softmax')
])
```

```
model.summary()
```

```
import tensorflow as tf

optimizer = tf.keras.optimizers.Adam(learning_rate=0.001*0.1)

model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])
```

```
history = model.fit(train_x, train_y, epochs=50, verbose=1, validation_split=0.1)
```

```
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.show()
```

```
loss, acc = model.evaluate(test_x, test_y)

print("loss :", loss)
print("acc :", acc)
```
