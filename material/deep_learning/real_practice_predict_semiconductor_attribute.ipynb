```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 데이터 읽기

```
df = pd.read_csv("train.csv")
```

```
data = df.to_numpy()
print(data.shape)
print(data[0])
```

# 데이터 x, y 로 분리

```
np.random.shuffle(data)

x, y = data[:,4:], data[:,:4]
print(x.shape)
print(y.shape)
```

# x 데이터 파악

```
for i in range(x.shape[1]):
  print(i)
  plt.hist(x[:,i])
  plt.show()
```

```
print(np.min(x), np.max(x))
```

# y 데이터 파악

```
for i in range(y.shape[1]):
  print(i)
  plt.hist(y[:,i])
  plt.show()
```

## nomalization

```
min = np.min(y)
max = np.max(y)
y = (y-min) / (max-min)
```

```
print(np.min(y), np.max(y))
```

# train, test 데이터 분리

```
split_index = int(len(x)*0.8)
train_x, test_x = x[:split_index], x[split_index:]
train_y, test_y = y[:split_index], y[split_index:]
```

# 모델 준비

```
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import optimizers
from tensorflow.keras.layers import *
```

```
print(train_x.shape)
print(train_x.shape[1])
```

```
model = keras.Sequential([
    Dense(200, activation='relu', input_shape=(train_x.shape[1],)),
    Dense(64, activation='relu'),
    Dense(train_y.shape[1])
])
    
```

```
model.summary()
```

```
model.compile(optimizer='adam',
              loss='mse',
              metrics=['mae'])
```

## 학습

```
history = model.fit(train_x, train_y, epochs=50, verbose=1, batch_size=1024, validation_split=0.1)
```

```
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```

```
loss, mae = model.evaluate(test_x, test_y, batch_size=1024)

print("loss :", loss)
print("mae :", mae)
```

```
y_ = model.predict(test_x)
print(y_[:10])
```

# Classification 으로

## for layer_1

```
raw_x, raw_y = data[:,4:], data[:,:4]
```

```
y = raw_y[:,0]
# y = 10 ~ 300

y = (y/10.0-1).astype(int)
# y = 0 ~ 29
```

```
print(y.shape)
print(np.min(y), np.max(y))
print(y[:10])
```

```
split_index = int(len(x)*0.8)
train_x, test_x = x[:split_index], x[split_index:]
train_y, test_y = y[:split_index], y[split_index:]
```

```
train_x = np.expand_dims(train_x, -1)
test_x = np.expand_dims(test_x, -1)
```

```
from tensorflow.keras.layers import *

model = keras.Sequential([
    Conv1D(32, 9, activation='relu', padding="same", input_shape=(train_x.shape[1],1)),
    MaxPooling1D(2),
    Conv1D(32, 9, activation='relu', padding="same"),
    MaxPooling1D(2),
    Flatten(),
    Dense(200, activation='relu'),
    Dense(64, activation='relu'),
    Dense(30, activation="softmax")
])
```

```
model.compile(optimizer="SGD", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```
history = model.fit(train_x, train_y, epochs=50, verbose=1, batch_size=1024, validation_split=0.1)
```

```
loss, acc = model.evaluate(test_x, test_y)

print("loss :", loss)
print("acc :", acc)
```

```
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```

## MAE 계산

```
y_ = model.predict(test_x)
prediction = np.argmax(y_, axis=1)

diff = ((test_y - prediction)*10)

mae = np.mean(np.abs(diff))
print(mae)
```

```
plt.hist(diff, bins=300)
plt.show()
```
